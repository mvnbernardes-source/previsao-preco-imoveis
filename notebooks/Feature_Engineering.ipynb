```python
import pandas as pd
```

```python
base = pd.read_csv('../data/raw/house-price.csv')
base.head()
```

```python
print(base.columns)
```

```python
base.isnull().sum()
```

```python
# Separar variávvel alvo
y = base['price']
```

```python
# Separar variáveis explicativas
X = base.drop(columns=['price'], axis=1)
```

```python
# Transformar variáveis categóricas em numéricas
X = pd.get_dummies(X, drop_first=True).astype(int)
```

```python
# Juntar X e y para salvar o dataset completo
base_tratada = pd.concat([X, y], axis=1)
base_tratada.info()
```

```python
# Salvar a base tratada
base_tratada.to_csv('../data/processed/base_tratada.csv', index=False)
```
