```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('../data/raw/house-price.csv')
```

```python
base.head()
```

```python
base.info()
base.describe()
```

```python
base.isnull().sum().sort_values(ascending=False)
```

```python
# Histograma do preço
sns.histplot(base['price'], bins=50, kde=True)
plt.title('Distribuição do Preço dos Imóveis')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.savefig('../imagens/preco_histograma.png')
plt.show()
```

```python
# Correlação entre area e preço
sns.scatterplot(x='area', y='price', data=base)
plt.title('Correlação entre Área e Preço')
plt.xlabel('Área (m²)')
plt.ylabel('Preço')
plt.savefig('../imagens/correlacao_area_preco.png')
plt.show()
```

```python
# Correlação
plt.figure(figsize=(12, 10))
sns.heatmap(base.select_dtypes(include=[np.number]).corr(), cmap='coolwarm')
plt.title('Mapa de Correlação')
plt.savefig('../imagens/mapa_correlacao.png')
plt.show()
```
