```python
import pandas as pd
import joblib
```

```python
#carregar o modelo de regressão linear
modelo_lr = joblib.load('../modelo/modelo_regressao_linear.pkl')
```

```python
novos_imoveis = pd.read_csv('../data/predict/novos_imoveis.csv')
novos_imoveis.head()
```

```python
# Refazer a ordem das colunas para corresponder ao modelo
novos_imoveis = novos_imoveis[modelo_lr.feature_names_in_]
```

```python
# Prever os preços dos novos imóveis
previsoes = modelo_lr.predict(novos_imoveis)
print(previsoes)
```

```python
# Adicionar as previsões ao DataFrame
previsoes_novos_imoveis = pd.DataFrame(previsoes, columns=['preco_previsto'])
previsoes_novos_imoveis = previsoes_novos_imoveis["preco_previsto"].astype(int)
```

```python
previsoes_novos_imoveis.head()
```

```python
# adicionar as previsões ao DataFrame original
novos_imoveis.loc[:,'preco_previsto'] = previsoes_novos_imoveis
novos_imoveis.head()
```

```python
# Salvando o dataframe com as previsões em um novo arquivo CSV
novos_imoveis.to_csv('../data/predict/previsoes_imoveis.csv', index=False)
```
